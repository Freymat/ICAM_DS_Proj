```python
##### !!!!!  Ne pas effacer !!!!! #######
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Charger les données textuelles des ingrédients
ingredient_texts = df['ingredients_text']

# Paramètres
vocab_size = 5000 # nombre de mots uniques que le modèle va considérer lors de l'apprentissage, parmis les mots les plus fréquents.
max_length = 100 # longueur maximale des séquences
embedding_dim = 200

# Tokenisation des textes des ingrédients
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(ingredient_texts)
sequences = tokenizer.texts_to_sequences(ingredient_texts)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')

## Visualisation des informations sur les données d'ingrédients
# Longueur de la liste d'ingrédients la plus longue
max_list_length = max(len(seq) for seq in sequences)

# Longueur de la liste d'ingrédients la plus courte
min_list_length = min(len(seq) for seq in sequences)

# Longueur moyenne des listes d'ingrédients
avg_length = sum(map(len, sequences)) / len(sequences)

print("Longueur de la liste d'ingrédients la plus longue:", max_list_length)
print("Longueur de la liste d'ingrédients la plus courte:", min_list_length)
print("Longueur moyenne des listes d'ingrédients:", avg_length)

# Affichage d'une liste d'ingrédients et de la valeur de la colonne df['gluten_presence'] associée
index_to_display = 0
print("Liste d'ingrédients:", ingredient_texts[index_to_display])
print("Valeur de la colonne 'gluten_presence' associée:", df['gluten_presence'][index_to_display])

# Affichage de la même liste d'ingrédients après tokenisation/sequencing et padding
print("Séquence tokenisée/padded:", padded_sequences[index_to_display])
print("\n")

# Répartition des valeurs de la colonne 'gluten_presence'
print("Répartition des valeurs de la colonne 'gluten_presence':")
print(df['gluten_presence'].value_counts())


# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['gluten_presence'], test_size=0.2, random_state=42)

# Définition du modèle
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.LSTM(128, return_sequences=True),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Définition des métriques
METRICS = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
    tf.keras.metrics.AUC(name='auc'),
]

# Compilation du modèle avec les métriques définies
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=METRICS)

# Display a summary of the models structure
model.summary()

# Entraînement du modèle sur les données d'entraînement
model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=128)
```

```python
##### !!!!!  Ne pas effacer !!!!! #######
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Charger les données textuelles des ingrédients
ingredient_texts = df['ingredients_text']

# Paramètres
vocab_size = 5000 # nombre de mots uniques que le modèle va considérer lors de l'apprentissage, parmis les mots les plus fréquents.
max_length = 100 # longueur maximale des séquences
embedding_dim = 200

# Tokenisation des textes des ingrédients
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(ingredient_texts)
sequences = tokenizer.texts_to_sequences(ingredient_texts)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')

## Visualisation des informations sur les données d'ingrédients
# Longueur de la liste d'ingrédients la plus longue
max_list_length = max(len(seq) for seq in sequences)

# Longueur de la liste d'ingrédients la plus courte
min_list_length = min(len(seq) for seq in sequences)

# Longueur moyenne des listes d'ingrédients
avg_length = sum(map(len, sequences)) / len(sequences)

print("Longueur de la liste d'ingrédients la plus longue:", max_list_length)
print("Longueur de la liste d'ingrédients la plus courte:", min_list_length)
print("Longueur moyenne des listes d'ingrédients:", avg_length)

# Affichage d'une liste d'ingrédients et de la valeur de la colonne df['gluten_presence'] associée
index_to_display = 0
print("Liste d'ingrédients:", ingredient_texts[index_to_display])
print("Valeur de la colonne 'gluten_presence' associée:", df['gluten_presence'][index_to_display])

# Affichage de la même liste d'ingrédients après tokenisation/sequencing et padding
print("Séquence tokenisée/padded:", padded_sequences[index_to_display])
print("\n")

# Répartition des valeurs de la colonne 'gluten_presence'
print("Répartition des valeurs de la colonne 'gluten_presence':")
print(df['gluten_presence'].value_counts())


# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, df['gluten_presence'], test_size=0.2, random_state=42)

# Définition du modèle
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Définition des métriques
METRICS = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
    tf.keras.metrics.AUC(name='auc'),
]

# Compilation du modèle avec les métriques définies
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=METRICS)

# Display a summary of the models structure
model.summary()

# Entraînement du modèle sur les données d'entraînement
model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=128)
```
